```python
from primesense import openni2
from primesense import _openni2 as c_api
import numpy as np
import cv2
openni2.initialize("./Redist")
dev = openni2.Device.open_any()
print(dev.get_device_info())
depth_stream = dev.create_depth_stream()
depth_stream.start()

while True:
    # 显示深度图
    frame = depth_stream.read_frame()
    dframe_data = np.array(frame.get_buffer_as_triplet()).reshape([480, 640, 2])
    # 在不更改其数据的情况下为数组赋予新形状。
    # numpy.reshape(a, newshape, order='C')
    dpt1 = np.asarray(dframe_data[:, :, 0], dtype='float32')
    # array和asarray都可以将结构数据转化为ndarray，
    # 但是主要区别就是当数据源是ndarray时，array仍然会copy出一个副本，占用新的内存，但asarray不会。
    dpt2 = np.asarray(dframe_data[:, :, 1], dtype='float32')
    # 留空就全选
    # X[n,:]是取第1维中下标为n的元素的所有值。
    # X[1,:]即取第一维中下标为1的元素的所有值
    # X[:, m:n]，即取所有数据的第m到n-1列数据，含左不含右
    dpt2 *= 255
    dpt = dpt1 + dpt2
    cv2.imshow('dpt', dpt)
    # 按下q键退出循环
    key = cv2.waitKey(10)
    if int(key) == 113:
        break


depth_stream.stop()
dev.close()
```

```python
# 摄像头并显示轮廓
import cv2
cap = cv2.VideoCapture(0)
# 自己看电脑是数字几，这里是1调用内置的摄像头

i = 0
while(1):
    ret, frame = cap.read()
    img_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    img_gb = cv2.GaussianBlur(img_gray, (5, 5), 0)
    edges = cv2.Canny(img_gb, 100, 200)
    cv2.imshow("capture", edges)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
